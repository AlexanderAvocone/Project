```
import matplotlib.pyplot as plt
import numpy as np
"""Gegebene Funktionen """
def t1(r):return r**2
def t2(r):return np.exp(r)
def t3(r):return np.tan(r)
def t4(r):return np.log(1.+r)

""" Dichtefunktion == Ableitung der Umkehrfunktion """
def f1(x):return 1/(2*np.sqrt(x)) 
def f2(x):return 1/x 
def f3(x):return 1/(x**2+1)
def f4(x):return np.exp(x) 


n = np.random.random(100000)



""" Subplots """
plt.figure(figsize=(12,6))

#sub1
plt.subplot(2,2,1)
plt.hist(t1(n), bins = 20,density = True, alpha = 0.5)
x1 = np.linspace(0,1,100)
plt.plot(x1 , f1(x1),color = "r",alpha = 0.7)
plt.title("Dichtefunktion = 0.5/sqrt(x)")
plt.grid()

#sub2
plt.subplot(2,2,2)
x2 = np.linspace(1,np.e,100)
plt.hist(t2(n), bins = 20,density = True, alpha = 0.5)
plt.plot(x2, f2(x2),color = "r", alpha = 0.7)
plt.title("Dichtefunktion =  1/x")
plt.grid()

#sub3
plt.subplot(2,2,3)
x3 = np.linspace(0,1.55,150)
plt.hist(t3(n), bins = 20,density = True, alpha = 0.5)
plt.plot(x3,f3(x3),color = "r", alpha = 0.7)
plt.title("Dichtefunktion =  1/(x^2+1)")
plt.grid()


#sub4
plt.subplot(2,2,4)
x4 = np.linspace(0,0.68,100)
plt.hist(t4(n), bins = 20,density = True, alpha = 0.5)
plt.plot(x4,f4(x4),color = "r", alpha = 0.7)
plt.title("Dichtefunktion = e^x")
plt.grid()


plt.tight_layout()

```
