#**a) und b): Vergleich**



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp
from scipy.optimize import curve_fit

def binomial(x, n, p):
  k=np.around(x)
  return (sp.binom(n,k ) * p**k * (1.-p)**(n-k))

def gauss(x,a,mu,sigma):
    return (a*np.exp(-(x-mu)**2/2/sigma**2)/np.sqrt(2*np.pi)/sigma)





n = 1000 
N = 10 

i=1
verteilung = []

while i < n:
    
    verteilung.append(sum(np.random.randint(0,2,N)))
    i+=1

x1 = np.arange(0,10,0.1)

p = 0.5

plt.plot(x1,binomial(x1,N,p)*n ,color = "red", label = "Binomialverteilung")
y,x,useless  = plt.hist(verteilung, bins = np.arange(0.5,N+0.5,1), color = "lightblue",edgecolor = "b", label ="Experiment")
plt.xlabel("Anzahl an MÃ¼nzen mit Zahl")
plt.ylabel("HÃ¤ufigkeit der Ereignisse")
plt.legend()
plt.show()


```

#**c): Erwartungswert und Varianz**

```python

x = x+0.5
y = np.append(y,[0])

#curve_fit
variablen, matrix = curve_fit(gauss, x, y, p0 = [200,5,0.5])
plt.plot(x1, gauss(x1,*variablen), color = "black")
plt.scatter(x,y)
plt.grid()
plt.show()
print("Erwartungswert und Varianz von \nder Gaussfunktion und der Binomialverteilung:\n")
print("Erwartungswert   Gauss :",round(variablen[1],2),"|    Binomial:", 0.5*N )
print("Varianz          Gauss :", round(variablen[2]**2,2),"|    Binomial:", N*p*(1-p))
```
