##**a)**

```python
from __future__ import print_function, division, unicode_literals

import numpy as np
import matplotlib.pyplot as plt

def wuerfele(rnd,wuerfelzahl):
    return(sum(rnd)/wuerfelzahl)

def fgauss(x,mu,sigma):
    return (np.exp(-(x-mu)**2/2/sigma**2)/np.sqrt(2*np.pi)/sigma)


n = 3
N = 1000
i=1

w1 = (np.random.randint(1,7,N))
while i < n:
    w1 = w1 + (np.random.randint(1,7,N))
    i+=1
    

plt.hist(w1, bins = np.arange(0.5,7*n,1), edgecolor = "b")
plt.title("Aufgabe 4.1a)")
plt.ylabel("Häufigkeit")
plt.xlabel("Gesamtaugenzahl")
plt.grid()
```

##**b)**

```python
# Schritt 2: Bestimmung der Haeufigkeit:
j = 1
w2 = (np.random.randint(1,7,1000))

while j < 10:   
    w2 = w2 + (np.random.randint(1,7,1000))
    w10 = w2
    j+=1

while j < 50:   
    w2 = w2 + (np.random.randint(1,7,1000))
    w50 = w2
    j+=1

while j < 100:   
    w2 = w2 + (np.random.randint(1,7,1000))
    w100 = w2
    j+=1


h1 = np.zeros(60)
h2 = np.zeros(300)
h3 = np.zeros(600)

print("Mittlere Augenzahl für n = 10:",int(wuerfele(w10,10)))
print("Mittlere Augenzahl für n = 50:",int(wuerfele(w50,50)))
print("Mittlere Augenzahl für n = 100:",int(wuerfele(w100,100)))

for zahl in w10:
    #Zahl = die Zahl an der stelle w2[i]
    h1[zahl-1] += 1

for zahl in w50:
    #Zahl = die Zahl an der stelle w2[i]
    h2[zahl-1] += 1
for zahl in w100:
    #Zahl = die Zahl an der stelle w2[i]
    h3[zahl-1] += 1


print("Häufigkeit von h10 :",max(h1))
print("Häufigkeit von h50 :",max(h2))   
print("Häufigkeit von h100:",max(h3))


plt.bar((int(wuerfele(w10,10)),int(wuerfele(w50,50)),int(wuerfele(w100,100))),[max(h1),max(h2),max(h3)] , edgecolor = "b")
plt.title("Aufgabe 4.1b)")
plt.ylabel("Häufigkeit")
plt.xlabel("Mittlere Augenzahl")
plt.show()
```

##**c)**

```python
m = 100
M = 10000
k = 1
w3 = (np.random.randint(1,7,M))
while k < m:
    w3 = w3 + (np.random.randint(1,7,M))
    k+=1


mu = wuerfele(w3,M)
sigma = w3.std()
print("Gaussverteilung: \n    mu =",round(mu,2),"\n sigma =",round(sigma,2))




plt.hist(w3, bins =np.arange(200,500,1), edgecolor = "black", alpha = .2)
plt.title("Aufgabe c)")
plt.ylabel("Häufigkeit")
plt.xlabel("Gesamtaugenzahl")


x = np.arange(200., 500., 0.1)
plt.plot(x, fgauss(x,mu,sigma)*(len(w3)), 'r-') 


plt.show()
```
